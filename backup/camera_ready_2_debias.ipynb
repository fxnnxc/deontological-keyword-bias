```python
import os 
import matplotlib.pyplot as plt
import seaborn as sns 
import pandas as pd
experiments = [
    'exp_6_1_debising',
    'exp_6_2_debising', 
]

models = [
    'openai--gpt-4o',
    'openai--gpt-4o-mini', 
    'llama3_1_instruct--70b',
    'llama3_1--8b',
    'gemma2--9b',
    'qwen2--7b',
    'exaone--8b',
]

promt_types = [
    'General',
    'Explicit',
    'Strict'
] 
answer_formats = [
    'Binary',
    'Regression'
]

results = {
    'experiment': [],
    'model': [],
    'N+': [],
    'N-': [],
    'positive_proportion': [],
    'modal_expression': []
}
logic_results = {
    'experiment': [],
    'model': [],
    'N+': [],
    'N-': [],
    'positive_proportion': [],
    'modal_expression': []
}
for experiment in experiments:
    for model in models:
        for n_plus in [0,1,2]:
            for n_minus in [0,1,2]:
                csv_path = f"results/{experiment}/fewshot_{n_plus}_{n_minus}/analysis_summary_{model}.csv"
                if os.path.exists(csv_path):
                    df = pd.read_csv(csv_path)
                    
                    positive_proportion_with_me = df[df['sentence_type'] == 'strong']['ones_ratio'].values[0]
                    # positive_proportion_without_me = df[df['sentence_type'] == 'none']['ones_ratio'].values[0]
                    
                    # for modal_expression, positive_proportion in [('with', positive_proportion_with_me), ('without', positive_proportion_without_me)]:
                    for modal_expression, positive_proportion in [('with', positive_proportion_with_me)]:
                        results['experiment'].append(experiment)
                        results['model'].append(model)
                        results['N+'].append(n_plus)
                        results['N-'].append(n_minus)
                        results['positive_proportion'].append(positive_proportion)
                        results['modal_expression'].append(modal_expression)
        
        # 2 - 2 - Logic 
        csv_path = f"results/{experiment}/logicshot_2_2/analysis_summary_{model}.csv"
        if os.path.exists(csv_path):
            df = pd.read_csv(csv_path)
            positive_proportion_with_me = df[df['sentence_type'] == 'strong']['ones_ratio'].values[0]
            # positive_proportion_without_me = df[df['sentence_type'] == 'none']['ones_ratio'].values[0]
            
            for modal_expression, positive_proportion in [('with', positive_proportion_with_me)]:
                logic_results['experiment'].append(experiment)
                logic_results['model'].append(model)
                logic_results['N+'].append(2)
                logic_results['N-'].append(2)
                logic_results['positive_proportion'].append(positive_proportion)
                logic_results['modal_expression'].append(modal_expression)

dfs_logic = pd.DataFrame(logic_results)
dfs = pd.DataFrame(results)
dfs
```
